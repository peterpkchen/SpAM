```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
db_shp=gpd.read_file('Data/DB_shp')
cd_shp=gpd.read_file('Data/CD_shp')
```

```python
cd_list=cd_shp[['CDUID', 'CDNAME', 'PRUID']]
cd_list=cd_list[cd_list['PRUID'].astype(int)==35]
cd_list
```

```python
on_db_shp=db_shp[db_shp['PRUID'].astype(int)==35]
```

```python
on_db_shp['CDUID']=on_db_shp['DBUID'].astype(str).str[0:4]
```

```python
on_db_shp.merge(cd_list, on='CDUID')
on_db_shp.plot()
```
